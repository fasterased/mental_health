```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Загрузка датасета
df = pd.read_csv('/content/Mental_Health_Lifestyle_Dataset (1).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
missing_values = df.isnull().sum()
print(missing_values)

```

```python
df = df.dropna()
```

```python
numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Корреляции между числовыми признаками")
plt.show()
```

```python
numerical_df = df.select_dtypes(include=['number'])

# Построение графиков распределения
for column in numerical_df.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(numerical_df[column], bins=20, color='blue')
    plt.title(f'Распределение: {column}')
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.show()

```

```python
#простая описательная статистика
descriptive_stats = numerical_df.describe()
print(descriptive_stats)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Exercise Level', y='Happiness Score', hue='Diet Type')
plt.title('Уровень счастья в зависимости от физической активности и диеты')
plt.xlabel('Уровень физической активности')
plt.ylabel('Уровень счастья')
plt.legend(title='Тип диеты')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Stress Level', y='Screen Time per Day (Hours)')
plt.title('Экранное время в зависимости от уровня стресса')
plt.xlabel('Уровень стресса')
plt.ylabel('Экранное время (часов в день)')
plt.show()
```
